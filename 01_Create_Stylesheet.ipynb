# Tips and tricks for using matplotlib

## Make a stylesheet file
To change the defaults for plots created with matplotlib, it is most convenient to create a stylesheet file, which you can simply load after importing matplotlib. It may also be a good idea to create different style sheets, i.e. for your PhD thesis, for papers, and so on.

To start, simply add a stylesheet in your matplotlib config directory as `stylelib/STYLENAME.mplstyle`. Don't forget to create the stylelib folder if it does not exist. You can locate your config directory through the following command:

````python
import matplotlib
print(matplotlib.get_configdir())
````

In this stylesheet file, you can set the default for all `matplotlib.RcParams`. For reference, you can find the content of my file `C:\Users\andre\.matplotlib\stylelib\thesis.mplstyle` below, with some comments added.

```INI
## Default figure size in inches. For me this gave a good font size in the PhD thesis (booklet) format
figure.figsize : 8,3.5     
figure.dpi : 120

## We do not want tight_layout, but constrained_layout. The first one simply crops the
## whitespace, which leads to different image sizes depending on colorbars etc.
figure.autolayout : False
figure.constrained_layout.use : True

savefig.dpi : 250
savefig.pad_inches : 0.1 # The whitespace around your images

## If you enable usetex, latex is called to do ALL text, which makes it very slow.
## However, matplotlib itself understands a lot of TEX commands, so there is no need to to use latex.
## I find stixsans to be a pleasant font for plots.
text.usetex : False
mathtext.fontset :  stixsans

axes.labelsize : 11

## Enable smooth boxes around the legend
legend.fancybox : True
```

Once you created your personal stylesheet, you can enable it by executing `plt.style.use('STYLENAME')`. Let's have a look at the default

````python
import matplotlib.pyplot as plt

plt.plot(range(5),[0,4,1,2,8])
plt.plot(range(5),[3,0,2,7,6])
plt.title(r"Default style $\pi \omega_0 \Psi$ ")
````

And now the same with the stylesheet enabled:

````python
plt.style.use('thesis')
````

````python
plt.plot(range(5),[0,4,1,2,8])
plt.plot(range(5),[3,0,2,7,6])
plt.title(r"Thesis style $\pi \omega_0 \Psi$ ")
````

Keep in mind that "using" a stylesheet only changes the parameters that are mentioned in the stylesheet file while the others remain untouched. So if you use multiple stylesheets after each other, things might get messy. Either make sure that you have all relevant setting set (back) in all stylesheets you use or always reset the kernel to start with the defaults.
